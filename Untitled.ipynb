```python
# opencv画像入力用
from IPython.display import display, Image

def display_cv_image(image, format='.JPG'):
    decoded_bytes = cv2.imencode(format, image)[1].tobytes()
    display(Image(data=decoded_bytes))

def readyForGuess(img_name,width):
    img = cv2.imread( "./img/"+img_name+".jpg", cv2.IMREAD_GRAYSCALE)
    size = (3*width,4*width)
    #Create Small Size Image
    length=size[0]*size[1]
    img = cv2.resize(img, size)
    img_dat = img/255
    img_dat = img_dat.reshape(1,length)
    return img_dat

def showDiff(num):
    print("raw")
    display_cv_image(raw_img[num])
    print("processed")
    display_cv_image(proc_img[num])
    print("diff")
    display_cv_image(diff_img[num])
    
def resizeAndShow(img,num):
    size = (3*num,4*num)
    show_img = cv2.resize(img, size)
    display_cv_image(show_img)
    
    
def find_rect_of_target_color(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV_FULL)
    h = hsv[:, :, 0]
    s = hsv[:, :, 1]
    v = hsv[:, :, 2]
    return h,s,v
```

```python
#画像読み込みします
width = 120
size = (3*width,4*width)
length=size[0]*size[1]

raw_img = []
proc_img = []
eye_img = []


dir = os.getcwd()# カレントディレクトリのパスを取得


#positive
pos_dir = dir+"/img/raw/"
files = os.listdir(pos_dir)# ファイルのリストを取得
for file in files:# ファイルの数だけループ
    index = re.search('.JPG', file)# 拡張子がjpgのものを検出
    if index:# jpgの時だけ
        read_img = cv2.imread( pos_dir+file)
        read_img = cv2.resize(read_img, size)
        raw_img.append(read_img)
#negative
neg_dir = dir+"/img/processed/"
files = os.listdir(neg_dir)# ファイルのリストを取得
for file in files:# ファイルの数だけループ
    index = re.search('.JPG', file)# 拡張子がjpgのものを検出
    if index:# jpgの時だけ
        read_img = cv2.imread( neg_dir+file)
        read_img = cv2.resize(read_img, size)
        read_img = read_img
        proc_img.append(read_img)

#eye
eye_dir = dir+"/img/eye/"
files = os.listdir(eye_dir)# ファイルのリストを取得
for file in files:# ファイルの数だけループ
    index = re.search('.JPG', file)# 拡張子がjpgのものを検出
    if index:# jpgの時だけ
        read_img = cv2.imread( eye_dir+file)
        read_img = cv2.resize(read_img, size)
        read_img = read_img
        eye_img.append(read_img)




# arrayに変換
raw_img = np.asarray(raw_img)
proc_img = np.asarray(proc_img)
eye_img = np.asarray(eye_img)

# diff_img = raw_img*0 #真っ黒のボックスを作成
```

```python
# display_cv_image(proc_img[0])
face_cascade = cv2.CascadeClassifier('/usr/local/Cellar/opencv/3.3.0_3/share/OpenCV/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('/usr/local/Cellar/opencv/3.3.0_3/share/OpenCV/haarcascades/haarcascade_eye.xml')
# 顔を検知

proc_eye_rect = [(0,0),(0,0)]
proc_eye_size = [(0,0),(0,0)]
raw_eye_rect = [(0,0),(0,0)]
raw_eye_size = [(0,0),(0,0)]
num = 0

img = proc_img[num]
faces = face_cascade.detectMultiScale(img)
for (x,y,w,h) in faces:
    counter = 0
    # 検知した顔を矩形で囲む
#     cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    # 顔画像（グレースケール）
    roi_gray = img[y:y+h, x:x+w]
    # 顔画像（カラースケール）
    roi_color = img[y:y+h, x:x+w]
    # 顔の中から目を検知
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        # 検知した目を矩形で囲む
        if(counter>=2):
            break
#         cv2.rectangle(roi_color,(ex,ey+int(eh/4)),(ex+ew,ey+int(eh*3/4)),(0,255,0),2)
        proc_eye_rect[counter] = (ex+x,ey+y)
        proc_eye_size[counter] = (ew,eh)
        counter +=1
        
resizeAndShow(img,70)
print("proc_show")
print("左上:",proc_eye_rect)
print("幅",proc_eye_size)

img = raw_img[num]
faces = face_cascade.detectMultiScale(img)

for (x,y,w,h) in faces:
    counter = 0
    # 検知した顔を矩形で囲む
#     cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    # 顔画像（グレースケール）
    roi_gray = img[y:y+h, x:x+w]
    # 顔ｇ増（カラースケール）
    roi_color = img[y:y+h, x:x+w]
    # 顔の中から目を検知
    eyes = eye_cascade.detectMultiScale(roi_gray,scaleFactor=1.1)
    for (ex,ey,ew,eh) in eyes:
        # 検知した目を矩形で囲む
        if(counter>=2):
            break
#       cv2.rectangle(画像, 左上座標, 右下座標, 色, 線の太さ).
#       cv2.rectangle(roi_color,(ex,ey+int(eh/4)),(ex+ew,ey+int(eh*3/4)),(0,255,0),2)
        raw_eye_rect[counter] = (ex+x,ey+y)
        raw_eye_size[counter] = (ew,eh)
        counter +=1
resizeAndShow(img,70)    
print("raw_show")
print("左上:",raw_eye_rect)
print("幅",raw_eye_size)
```

```python
def overlayOnPart(src_image, overlay_image, posX, posY):

    # オーバレイ画像のサイズを取得
    ol_height, ol_width = overlay_image.shape[:2]

    # OpenCVの画像データをPILに変換
    #　BGRAからRGBAへ変換
    src_image_RGBA = cv2.cvtColor(src_image, cv2.COLOR_BGR2RGB)
    overlay_image_RGBA = cv2.cvtColor(overlay_image, cv2.COLOR_BGRA2RGBA)

    #　PILに変換
    src_image_PIL=pilim.fromarray(src_image_RGBA)
    overlay_image_PIL=pilim.fromarray(overlay_image_RGBA)

    # 合成のため、RGBAモードに変更
    src_image_PIL = src_image_PIL.convert('RGBA')
    overlay_image_PIL = overlay_image_PIL.convert('RGBA')

    # 同じ大きさの透過キャンパスを用意
    tmp = pilim.new('RGBA', src_image_PIL.size, (255, 255,255, 0))
    # 用意したキャンパスに上書き
    tmp.paste(overlay_image_PIL, (posX, posY), overlay_image_PIL)
    # オリジナルとキャンパスを合成して保存
    result = pilim.alpha_composite(src_image_PIL, tmp)

    return  cv2.cvtColor(np.asarray(result), cv2.COLOR_RGBA2BGRA)


# l_restore_color = cv2.cvtColor(l_restore, cv2.COLOR_RGB2GRAY)
# cv2.imwrite("l_trans.png",l_restore_color)
# trans = pilim.new('RGBA', l_restore_color.shape, (0, 0, 0, 0))
# trans = l_restore
# width = l_restore_color.size[0]
# height = l_restore_color.size[1]
# for x in xrange(width):
#     for y in xrange(height):
#         pixel = org.getpixel( (x, y) )
        
#         # 白なら処理しない
#         if pixel[0] == 255 and pixel[1] == 255 and pixel[2] == 255:
#             continue
        
#         # 白以外なら、用意した画像にピクセルを書き込み
#         trans.putpixel( (x, y), pixel )


l_trans = pilim.open("./l_trans.png")
l_trans =np.asarray(l_trans)
a =overlayOnPart(proc_img[0], l_trans, proc_eye_rect[0][0],proc_eye_rect[0][1]+10)
resizeAndShow(a,70)
a = cv2.cvtColor(a, cv2.COLOR_RGB2GRAY)
print(proc_eye_rect[0][0])
a[a<100] = 0
a[a>=100] = 255
resizeAndShow(a,70)


# trans = pilim.new('RGBA', l_restore.shape, (0, 0, 0, 0))
# trans = l_restore
# trans = pilim.fromarray(trans2)
# fuga = (proc_img[0].shape[0],proc_img[0].shape[1])
# trans2 = pilim.new('RGBA', fuga, (0, 0, 0, 0))
# trans2 = proc_img[0]
# trans2 = pilim.fromarray(trans2)
# # bg_pil = cv2.cvtColor(proc_img[0], cv2.COLOR_RGB2GRAY)
# # bg_pil = pilim.fromarray(bg_pil)

# trans2.paste(trans,(0,0),(255,255,255,0),trans)
```
