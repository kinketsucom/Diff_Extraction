```python


# display_cv_image(proc_img[0])
face_cascade = cv2.CascadeClassifier('/usr/local/Cellar/opencv/3.3.0_3/share/OpenCV/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('/usr/local/Cellar/opencv/3.3.0_3/share/OpenCV/haarcascades/haarcascade_eye.xml')

# 顔を検知
def extractEye(img):
    eye_rect = [(0,0),(0,0)]
    eye_size = [(0,0),(0,0)]
    faces = face_cascade.detectMultiScale(img)
    counter = 0
#     face_counter = 0
    for (x,y,w,h) in faces:
        print("face!")
#         face_counter += 1
        # 検知した顔を矩形で囲む
#         cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

        # 顔画像（グレースケール）
        roi_gray = img[y:y+h, x:x+w]
        # 顔画像（カラースケール）
        roi_color = img[y:y+h, x:x+w]
        # 顔の中から目を検知
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex,ey,ew,eh) in eyes:
            # 検知した目を矩形で囲む
            if(counter>=2):
                break
#             cv2.rectangle(roi_color,(ex,ey+int(eh/4)),(ex+ew,ey+int(eh*3/4)),(0,255,0),2)
            eye_rect[counter] = (ex+x,ey+y)
            eye_size[counter] = (ew,eh)
            counter +=1
    
    if(counter != 2):
        print("目が正確に検出できませんでした。",counter)

    
#     print("顔の個数:",face_counter)
    if (eye_rect[0][0]<eye_rect[1][0]):
        return eye_rect,eye_size
    else:
#         print("左右変更")
        tmp_rect = eye_rect[1]
        tmp_size = eye_size[1]
        eye_rect[1] = eye_rect[0]
        eye_rect[0] = tmp_rect
        eye_size[1] = eye_size[0]
        eye_size[0] = tmp_size
        return eye_rect,eye_size

def  extractLeftEyeForAI(eye_rect,eye_size,img):
    ex = eye_rect[0][0]
    ey = eye_rect[0][1]
    ew = eye_size[0][0]
    eh = eye_size[0][1]
    return img[ey+int(eh/4):ey+int(eh*3/4),ex:ex+ew] #配列的にはたて、よこ

def  extractRightEyeForAI(eye_rect,eye_size,img):
    ex = eye_rect[1][0]
    ey = eye_rect[1][1]
    ew = eye_size[1][0]
    eh = eye_size[1][1]
    return img[ey+int(eh/4):ey+int(eh*3/4),ex:ex+ew] #配列的にはたて、よこ

for i in range(0,num):
    print(i,"マイ目のデータを処理")
#     resizeAndShow(proc_img[i],70)
    display_cv_image(proc_img[i])
    # 目領域の取得
    proc_eye_rect,proc_eye_size = extractEye(proc_img[i])
    raw_eye_rect,raw_eye_size = extractEye(raw_img[i])
    #目領域を抽出
    left_ai_proc_img.append(extractLeftEyeForAI(proc_eye_rect,proc_eye_size,proc_img[i]))
    left_ai_raw_img.append(extractLeftEyeForAI(raw_eye_rect,raw_eye_size,raw_img[i]))
    right_ai_proc_img.append(extractRightEyeForAI(proc_eye_rect,proc_eye_size,proc_img[i]))
    right_ai_raw_img.append(extractRightEyeForAI(raw_eye_rect,raw_eye_size,raw_img[i]))

#     print("proc_l",i),display_cv_image(left_ai_proc_img[i])
#     print("raw_l",i),display_cv_image(left_ai_raw_img[i])
#     print("proc_r",i),display_cv_image(right_ai_proc_img[i])
#     print("raw_r",i),display_cv_image(right_ai_raw_img[i])

    #サイズは適当です
    size = (100,50)

    #データ処理用にリサイズ
    l_proc_dat.append(cv2.resize(left_ai_proc_img[i], size))
    l_raw_dat.append(cv2.resize(left_ai_raw_img[i],size))
    r_proc_dat.append(cv2.resize(right_ai_proc_img[i], size))
    r_raw_dat.append(cv2.resize(right_ai_raw_img[i],size))
    
#     l_proc_dat[i] = np.asarray(l_proc_dat[i])   
#     l_raw_dat[i] = np.asarray(l_raw_dat[i])
#     r_proc_dat[i] = np.asarray(r_proc_dat[i])
#     r_raw_dat[i] = np.asarray(r_raw_dat[i])
#     #グレースケールに変換
#     l_g_proc_dat.append(cv2.cvtColor(l_proc_dat[i], cv2.COLOR_RGB2GRAY))
#     l_g_raw_dat.append(cv2.cvtColor(l_raw_dat[i], cv2.COLOR_RGB2GRAY))
#     r_g_proc_dat.append(cv2.cvtColor(r_proc_dat[i], cv2.COLOR_RGB2GRAY))
#     r_g_raw_dat.append(cv2.cvtColor(r_raw_dat[i], cv2.COLOR_RGB2GRAY))
#     #2値化けのためにnp配列に変換
#     l_g_proc_dat[i] = np.asarray(l_g_proc_dat[i])
#     l_g_raw_dat[i] = np.asarray(l_g_raw_dat[i])
#     r_g_proc_dat[i] = np.asarray(r_g_proc_dat[i])
#     r_g_raw_dat[i] = np.asarray(r_g_raw_dat[i])
#     # 閾値
#     t = 80
#     l_g_proc_dat[i][l_g_proc_dat[i]<t] = 0
#     l_g_proc_dat[i][l_g_proc_dat[i]>=t] = 1
#     l_g_raw_dat[i][l_g_raw_dat[i]<t] = 0
#     l_g_raw_dat[i][l_g_raw_dat[i]>=t] = 1
#     r_g_proc_dat[i][r_g_proc_dat[i]<t] = 0
#     r_g_proc_dat[i][r_g_proc_dat[i]>=t] = 1
#     r_g_raw_dat[i][r_g_raw_dat[i]<t] = 0
#     r_g_raw_dat[i][r_g_raw_dat[i]>=t] = 1
    
#     display_cv_image(l_g_proc_dat[i]*255)
#     display_cv_image(l_g_raw_dat[i]*255)
#     display_cv_image(r_g_proc_dat[i]*255)
#     display_cv_image(r_g_raw_dat[i]*255)
```

# モデル形成

```python
l_model = Sequential()
l_model.add(Conv2D(3, (10,10), padding='same',data_format="channels_last" ,input_shape=l_size ))
# l_model.add(Dense((50,100), input_shape= (50,100) ))
# l_model.add(Activation('sigmoid'))
l_model.add(Activation('relu'))
# l_model.add(Dense(3))
# コンパイル
l_model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
l_model.summary()

r_model = Sequential()
r_model.add(Conv2D(3, (10,10), padding='same',data_format="channels_last" ,input_shape=r_size))
# r_model.add(Dense(1, input_shape= (2,50,100) ))
# r_model.add(Activation('sigmoid'))
r_model.add(Activation('relu'))
# r_model.add(Dense(3))
# コンパイル
r_model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
r_model.summary()
```
